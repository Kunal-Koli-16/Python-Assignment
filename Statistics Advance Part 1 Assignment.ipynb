```python
"""
#Q1 What is a random variable in probability theory?
A random variable is a numerical outcome of a random process. It assigns values to events in a sample space.
- Example: Tossing a coin → Random variable X can be defined as X=1 if heads, X=0 if tails.

===========================================================================================================
#Q2 What are the types of random variables?
- Discrete random variable: Takes countable values (e.g., number of heads in 10 coin tosses).
- Continuous random variable: Takes values from an interval of real numbers (e.g., height, weight).

===========================================================================================================
#Q3 What is the difference between discrete and continuous distributions?
- Discrete distribution: Probability is assigned to specific values.
- Continuous distribution: Probability is spread over intervals; individual points have probability 0.

===========================================================================================================
#Q4 What are probability distribution functions (PDF)?
- For discrete variables: Probability Mass Function (PMF) gives probability of each value.
- For continuous variables: Probability Density Function (PDF) describes likelihood density; area under curve = probability.
                                                                                                             
===========================================================================================================                                                                                                             
#Q5 How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- PDF/PMF: Probability at a point (or density).
- CDF: Probability that random variable X\leq x. It accumulates probabilities.

===========================================================================================================
#Q6 What is a discrete uniform distribution?
Each outcome in a finite set has equal probability.
- Example: Rolling a fair die → probability of each face = 1/6.

===========================================================================================================
#Q7 What are the key properties of a Bernoulli distribution?
- Two outcomes: success (1) with probability p, failure (0) with probability 1-p.
- Mean = p, Variance = p(1-p).

===========================================================================================================
#Q8 What is the binomial distribution, and how is it used in probability?
- Models number of successes in n independent Bernoulli trials.
- Parameters: n (trials), p (success probability).
- Example: Probability of 3 heads in 5 coin tosses.

===========================================================================================================
#Q9 What is the Poisson distribution and where is it applied?
- Models number of events in fixed time/space when events occur independently at constant average rate \lambda .
- Applications: Number of calls at a call center per hour, number of accidents per day.

===========================================================================================================
#Q10 What is a continuous uniform distribution?
- All values in interval [a,b] equally likely.
- PDF = 1/(b-a) for a\leq x\leq b.

===========================================================================================================
#Q11 What are the characteristics of a normal distribution?
- Bell-shaped, symmetric about mean.
- Mean = Median = Mode.
- Defined by mean \mu  and variance \sigma ^2.

===========================================================================================================
#Q12 What is the standard normal distribution, and why is it important?
- Normal distribution with mean 0 and variance 1.
- Used for standardization (Z-scores) and statistical inference.

===========================================================================================================
#Q13 What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- States that the sampling distribution of the sample mean approaches a normal distribution as sample size increases,
    regardless of population distribution.
- Critical because it justifies using normal-based methods in statistics.

===========================================================================================================
#Q14 How does the Central Limit Theorem relate to the normal distribution?
- CLT explains why normal distribution appears in practice: averages of large samples tend to be normally distributed.

===========================================================================================================
#Q15 What is the application of Z statistics in hypothesis testing?
- Z-statistic measures how far a sample mean is from population mean in standard deviation units.
- Used in tests for population mean when variance is known.

===========================================================================================================
#Q16 How do you calculate a Z-score, and what does it represent?
Z=\frac{X-\mu }{\sigma }
- Represents how many standard deviations a value X is from mean \mu .

===========================================================================================================
#Q17 What are point estimates and interval estimates in statistics?
- Point estimate: Single value estimate of a parameter (e.g., sample mean).
- Interval estimate: Range of values (confidence interval) likely to contain parameter.

===========================================================================================================
#Q18 What is the significance of confidence intervals in statistical analysis?
- Provide range of plausible values for population parameter.
- Reflect uncertainty in estimation.

===========================================================================================================
#Q19 What is the relationship between a Z-score and a confidence interval?
- Confidence intervals are constructed using Z-scores (critical values).
- Example: 95% CI uses Z = ±1.96.

===========================================================================================================
#Q20 How are Z-scores used to compare different distributions?
- Standardize values across different distributions.
- Allow comparison on a common scale.

===========================================================================================================
#Q21 What are the assumptions for applying the Central Limit Theorem?
- Independent and identically distributed (i.i.d.) samples.
- Finite mean and variance.
- Large sample size.

===========================================================================================================
#Q22 What is the concept of expected value in a probability distribution?
- Weighted average of all possible values.
E[X]=\sum x\cdot P(x)\quad \mathrm{(discrete)}\quad \mathrm{or}\quad E[X]=\int xf(x)dx\quad \mathrm{(continuous)}


===========================================================================================================
#Q23 How does a probability distribution relate to the expected outcome of a random variable?
- The distribution describes all possible outcomes and their probabilities.
- Expected value summarizes the "average" outcome over many trials.
"""
```

```python
#Q15 Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
import numpy as np
import matplotlib.pyplot as plt

sample_means = []
for _ in range(1000):
    sample = np.random.normal(0, 1, 30)  # Normal distribution
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, density=True, edgecolor='black')
plt.title("CLT Verification (Normal Distribution)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()



#Q16 Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal():
    x = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x, 0, 1)
    plt.plot(x, pdf)
    plt.title("Standard Normal Distribution (mean=0, std=1)")
    plt.xlabel("Z")
    plt.ylabel("Density")
    plt.show()

plot_standard_normal()



#Q17 Generate random variables and calculate their corresponding probabilities using the binomial distribution
import numpy as np
from scipy.stats import binom

n, p = 10, 0.5
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

for val, prob in zip(x, pmf):
    print(f"P(X={val}) = {prob:.4f}")



#Q18 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
from scipy.stats import norm

data_point = 75
mean = 70
std = 5

z_score = (data_point - mean) / std
print("Z-score:", z_score)

# Compare with standard normal distribution
prob = norm.cdf(z_score)
print("Probability of being <= data_point:", prob)



#Q19 Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

data = np.random.normal(50, 10, 100)
sample_mean = np.mean(data)
population_mean = 52
std = np.std(data, ddof=1)
n = len(data)

z_stat = (sample_mean - population_mean) / (std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Z-statistic:", z_stat)
print("p-value:", p_value)



#Q20 Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy.stats import norm

data = np.random.normal(100, 15, 50)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

confidence_level = 0.95
z = norm.ppf(1 - (1-confidence_level)/2)

margin_error = z * (std / np.sqrt(n))
ci_lower = mean - margin_error
ci_upper = mean + margin_error

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")



#Q21 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
from scipy.stats import norm

data = np.random.normal(50, 10, 100)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

confidence_level = 0.99
z = norm.ppf(1 - (1-confidence_level)/2)

margin_error = z * (std / np.sqrt(n))
ci_lower = mean - margin_error
ci_upper = mean + margin_error

print(f"99% Confidence Interval for mean: ({ci_lower:.2f}, {ci_upper:.2f})")



#Q22 Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, 0, 1)

plt.plot(x, pdf)
plt.title("PDF of Normal Distribution (mean=0, std=1)")
plt.xlabel("x")
plt.ylabel("Density")
plt.show()



#Q23 Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lambda_val = 4
x = np.arange(0, 15)
cdf = poisson.cdf(x, lambda_val)

plt.step(x, cdf, where='mid')
plt.title("CDF of Poisson Distribution (λ=4)")
plt.xlabel("Events")
plt.ylabel("Cumulative Probability")
plt.show()



#Q24 Simulate a random variable using a continuous uniform distribution and calculate its expected value
import numpy as np

samples = np.random.uniform(2, 8, 10000)
expected_value = np.mean(samples)

print("Expected Value (approx):", expected_value)




```

```python
#Q25 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
import numpy as np

# Generate dataset from normal distribution
data = np.random.normal(100, 20, 1000)

# Range
range_val = np.max(data) - np.min(data)

# Interquartile Range (IQR)
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1

print("Range:", range_val)
print("Interquartile Range (IQR):", iqr)



#Q26 Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt

# Generate dataset
data = np.random.normal(50, 10, 100)

# Z-score normalization
z_normalized = (data - np.mean(data)) / np.std(data)

# Visualization
plt.hist(z_normalized, bins=30, edgecolor='black')
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()



#Q27 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution
import numpy as np
from scipy.stats import skew, kurtosis

def skewness_kurtosis(data):
    return skew(data), kurtosis(data)

# Generate dataset
data = np.random.normal(0, 1, 1000)

skew_val, kurt_val = skewness_kurtosis(data)
print("Skewness:", skew_val)
print("Kurtosis:", kurt_val)




```
