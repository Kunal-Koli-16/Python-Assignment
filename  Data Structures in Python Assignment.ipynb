```python
#1. Reverse a string
def reverse_string(s):
    return s[::-1]

print(reverse_string("hello")) 
#==============================================================================

#2. Count vowels in a string
def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for ch in s if ch in vowels)

print(count_vowels("Programming"))  

#==============================================================================
#3. Check if a string is a palindrome
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome("madam")) 

#==============================================================================
#4. Check if two strings are anagrams
def are_anagrams(s1, s2):
    return sorted(s1) == sorted(s2)

print(are_anagrams("listen", "silent")) 

#==============================================================================
#5. Find all occurrences of a substring
def find_occurrences(text, sub):
    positions = []
    start = 0
    while True:
        start = text.find(sub, start)
        if start == -1:
            break
        positions.append(start)
        start += 1
    return positions

print(find_occurrences("banana", "ana"))  

#==============================================================================
#6. String compression (basic run-length encoding)
def compress_string(s):
    compressed = ""
    count = 1
    for i in range(1, len(s)):
        if s[i] == s[i-1]:
            count += 1
        else:
            compressed += s[i-1] + str(count)
            count = 1
    compressed += s[-1] + str(count)
    return compressed

print(compress_string("aaabbc"))

#==============================================================================
#7. Check if a string has all unique characters
def has_unique_chars(s):
    return len(set(s)) == len(s)

print(has_unique_chars("python"))  

#==============================================================================
#8. Convert string to uppercase/lowercase
def convert_case(s):
    return s.upper(), s.lower()

print(convert_case("Hello"))  

#==============================================================================
#9. Count words in a string
def count_words(s):
    return len(s.split())

print(count_words("Python is fun"))  

#==============================================================================
#10. Concatenate two strings without +
def concat_strings(s1, s2):
    return "{}{}".format(s1, s2)

print(concat_strings("Hello", "World"))  



```

```python
#11. Remove all occurrences of an element
def remove_element(lst, elem):
    return [x for x in lst if x != elem]

print(remove_element([1, 2, 3, 2, 4], 2)) 

#==============================================================================
#12. Find the second largest number
def second_largest(lst):
    unique = list(set(lst))
    unique.sort()
    return unique[-2] if len(unique) >= 2 else None

print(second_largest([10, 20, 4, 45, 99]))  

#==============================================================================
#13. Count occurrences of each element
def count_elements(lst):
    counts = {}
    for item in lst:
        counts[item] = counts.get(item, 0) + 1
    return counts

print(count_elements([1, 2, 2, 3, 3, 3]))  

#==============================================================================
#14. Reverse a list in-place
def reverse_list(lst):
    left, right = 0, len(lst) - 1
    while left < right:
        lst[left], lst[right] = lst[right], lst[left]
        left += 1
        right -= 1
    return lst

print(reverse_list([1, 2, 3, 4])) 

#==============================================================================
#15. Remove duplicates while preserving order
def remove_duplicates(lst):
    seen = set()
    result = []
    for item in lst:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

print(remove_duplicates([1, 2, 2, 3, 1])) 

#==============================================================================
#16. Check if a list is sorted
def is_sorted(lst):
    return lst == sorted(lst) or lst == sorted(lst, reverse=True)

print(is_sorted([1, 2, 3])) 
print(is_sorted([3, 2, 1]))  

#==============================================================================
#17. Merge two sorted lists
def merge_sorted_lists(l1, l2):
    i, j = 0, 0
    merged = []
    while i < len(l1) and j < len(l2):
        if l1[i] < l2[j]:
            merged.append(l1[i])
            i += 1
        else:
            merged.append(l2[j])
            j += 1
    merged.extend(l1[i:])
    merged.extend(l2[j:])
    return merged

print(merge_sorted_lists([1, 3, 5], [2, 4, 6]))  

#==============================================================================
```

```python
#18. Intersection of two lists
def list_intersection(lst1, lst2):
    return [x for x in lst1 if x in lst2]

print(list_intersection([1, 2, 3, 4], [3, 4, 5, 6]))  

#==============================================================================

#19. Union of two lists without duplicates
def list_union(lst1, lst2):
    return list(set(lst1 + lst2))

print(list_union([1, 2, 3], [3, 4, 5]))  


#==============================================================================

#20. Shuffle a list without built-in shuffle
import random

def custom_shuffle(lst):
    for i in range(len(lst)-1, 0, -1):
        j = random.randint(0, i)
        lst[i], lst[j] = lst[j], lst[i]
    return lst

print(custom_shuffle([1, 2, 3, 4, 5]))  



```

```python
#21. Common elements in two tuples
def tuple_intersection(t1, t2):
    return tuple(set(t1) & set(t2))

print(tuple_intersection((1, 2, 3), (2, 3, 4)))  

#==============================================================================

#22. Concatenate two tuples
def concat_tuples(t1, t2):
    return t1 + t2

print(concat_tuples((1, 2), (3, 4)))  

#==============================================================================

#23. Prompt user for two sets and print intersection
def set_intersection():
    set1 = set(map(int, input("Enter first set (comma-separated): ").split(',')))
    set2 = set(map(int, input("Enter second set (comma-separated): ").split(',')))
    print("Intersection:", set1 & set2)
#==============================================================================

#24. Elements in first set but not in second
def difference_of_sets():
    set1 = set(input("Enter first set of strings (comma-separated): ").split(','))
    set2 = set(input("Enter second set of strings (comma-separated): ").split(','))
    print("Elements in first set but not in second:", set1 - set2)

#==============================================================================

#25. Union of two sets of characters
def union_of_char_sets():
    set1 = set(input("Enter first set of characters (comma-separated): ").split(','))
    set2 = set(input("Enter second set of characters (comma-separated): ").split(','))
    print("Union:", set1 | set2)

#==============================================================================

#26. Union, intersection, and difference of two sets
def set_operations():
    set1 = {1, 2, 3, 4}
    set2 = {3, 4, 5, 6}
    print("Union:", set1 | set2)
    print("Intersection:", set1 & set2)
    print("Difference (set1 - set2):", set1 - set2)

#==============================================================================

#27. Symmetric difference of two sets of strings
def symmetric_difference_sets():
    set1 = set(input("Enter first set of strings (comma-separated): ").split(','))
    set2 = set(input("Enter second set of strings (comma-separated): ").split(','))
    print("Symmetric Difference:", set1 ^ set2)

#==============================================================================

#28. Slice tuple within range of indices
def tuple_slice(t, start, end):
    return t[start:end]

print(tuple_slice((10, 20, 30, 40, 50), 1, 4))  

#==============================================================================

#29. Max and Min using tuple unpacking
def tuple_min_max(t):
    max_val, min_val = max(t), min(t)
    return max_val, min_val

print(tuple_min_max((5, 10, 3, 8)))  

#==============================================================================

#30. Count occurrences of element in tuple
def count_in_tuple(t, elem):
    return t.count(elem)

print(count_in_tuple((1, 2, 2, 3, 2), 2))  

```

```python
#31. Word frequency dictionary
def word_frequency(words):
    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1
    return freq

print(word_frequency(["apple", "banana", "apple", "orange", "banana"]))  


#==============================================================================

#32. Merge two dictionaries (sum values for common keys)
def merge_dicts(d1, d2):
    merged = d1.copy()
    for key, value in d2.items():
        merged[key] = merged.get(key, 0) + value
    return merged

print(merge_dicts({'a': 2, 'b': 3}, {'b': 4, 'c': 5}))  

#==============================================================================

#33. Access value in nested dictionary
def access_nested_dict(d, keys):
    for key in keys:
        if isinstance(d, dict) and key in d:
            d = d[key]
        else:
            return None
    return d

nested = {"a": {"b": {"c": 42}}}
print(access_nested_dict(nested, ["a", "b", "c"])) 
print(access_nested_dict(nested, ["a", "x"]))       
#==============================================================================

#34. Sort dictionary by values
def sort_dict_by_values(d, ascending=True):
    return dict(sorted(d.items(), key=lambda item: item[1], reverse=not ascending))

print(sort_dict_by_values({'a': 3, 'b': 1, 'c': 2}, ascending=True))  

#==============================================================================

#35. Invert dictionary (handle duplicate values)
def invert_dict(d):
    inverted = {}
    for key, value in d.items():
        inverted.setdefault(value, []).append(key)
    return inverted

print(invert_dict({'a': 1, 'b': 2, 'c': 1}))  

```
