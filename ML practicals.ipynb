# prac no 5

```python
def predict_car_speed(age):
    initial_speed = 100
    speed_decrease_peryear =2
    predicted_speed = initial_speed - speed_decrease_peryear * age
    
    if predicted_speed < 0:
        predicted_speed=0
    return predicted_speed
age_car =5
predicted_speed = predict_car_speed(age_car)
print(f"age of car {age_car} speed is {predicted_speed}")
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn.linear_model import LinearRegression

x = np.array([[1,1],[1,2],[1,3],[1,4]])
y = np.array([[1],[3],[5],[7]])

model = LinearRegression().fit(x,y)

print("coe : " , model.coef_)
print("inter" , model.intercept_)
plt.plot(x,y)
plt.show()
```

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt 

x,y = make_circles(n_samples= 500 , shuffle = True , noise =0.15 , random_state=42)

plt.scatter(x[:,0] , x[:,1] , c=y)
plt.show
```
